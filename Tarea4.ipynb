# Ejercicio 1 
Use el m√©todo de bisecci√≥n para encontrar soluciones precisas dentro de 10^-2 para ùë•
$$
10x^3 - 7x^2 + 14x - 6 = 0
$$

 en cada intervalo. 

a.[0,1]

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n objetivo
def funcion_objetivo(x):
    return x**3 - 7*x**2 + 14*x - 6

# M√©todo de Bisecci√≥n adaptado
def metodo_biseccion(funcion, inicio, fin, tolerancia=1e-2, iteraciones_max=100):
    if funcion(inicio) * funcion(fin) >= 0:
        print("Intervalo inv√°lido: el m√©todo de bisecci√≥n requiere signos opuestos en los extremos.")
        return None
    
    for i in range(iteraciones_max):
        punto_medio = (inicio + fin) / 2
        valor_medio = funcion(punto_medio)
        
        if abs(valor_medio) < tolerancia or (fin - inicio) / 2 < tolerancia:
            return punto_medio  # Se alcanza la tolerancia deseada o una ra√≠z exacta
        
        if funcion(inicio) * valor_medio < 0:
            fin = punto_medio
        else:
            inicio = punto_medio
    
    return (inicio + fin) / 2

# Definici√≥n del intervalo de b√∫squeda
intervalo_inicio, intervalo_fin = 0, 1

# B√∫squeda de la ra√≠z
raiz_aproximada = metodo_biseccion(funcion_objetivo, intervalo_inicio, intervalo_fin)

# Visualizaci√≥n de la funci√≥n y la ra√≠z aproximada
x_vals = np.linspace(intervalo_inicio, intervalo_fin, 500)
y_vals = funcion_objetivo(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = x^3 - 7x^2 + 14x - 6$", color='blue')
plt.axhline(0, color='gray', linewidth=0.7)
plt.axvline(raiz_aproximada, color='red', linestyle="--", label=f"Ra√≠z ‚âà {raiz_aproximada:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximaci√≥n de la ra√≠z usando el m√©todo de bisecci√≥n")
plt.legend()
plt.grid()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n a evaluar
def evaluar_funcion(x):
    return x**3 - 7*x**2 + 14*x - 6

# Funci√≥n para aplicar el m√©todo de bisecci√≥n
def encontrar_raiz_biseccion(func, limite_inf, limite_sup, tolerancia=1e-2, max_iter=100):
    if func(limite_inf) * func(limite_sup) >= 0:
        print("No es posible aplicar el m√©todo de bisecci√≥n en este intervalo.")
        return None
    
    for i in range(max_iter):
        punto_medio = (limite_inf + limite_sup) / 2
        valor_medio = func(punto_medio)
        
        if abs(valor_medio) < tolerancia or (limite_sup - limite_inf) / 2 < tolerancia:
            return punto_medio  # Soluci√≥n aproximada o exacta encontrada

        # Actualizar los l√≠mites en funci√≥n del cambio de signo
        if func(limite_inf) * valor_medio < 0:
            limite_sup = punto_medio
        else:
            limite_inf = punto_medio

    return (limite_inf + limite_sup) / 2  # Retorna la soluci√≥n aproximada

# Definir el intervalo para la b√∫squeda
inicio_intervalo, fin_intervalo = 3.2, 4

# Ejecutar el m√©todo de bisecci√≥n para encontrar la ra√≠z
raiz_encontrada = encontrar_raiz_biseccion(evaluar_funcion, inicio_intervalo, fin_intervalo)

# Preparar datos para graficar la funci√≥n en el intervalo especificado
x_datos = np.linspace(inicio_intervalo, fin_intervalo, 500)
y_datos = evaluar_funcion(x_datos)

# Generar la gr√°fica de la funci√≥n y marcar la ra√≠z encontrada
plt.figure(figsize=(8, 6))
plt.plot(x_datos, y_datos, label=r"$f(x) = x^3 - 7x^2 + 14x - 6$", color='teal')
plt.axhline(0, color='gray', linewidth=0.5)

# Marcar la ra√≠z en la gr√°fica si existe
if raiz_encontrada is not None:
    plt.axvline(raiz_encontrada, color='orange', linestyle="--", label=f"Ra√≠z ‚âà {raiz_encontrada:.2f}")
else:
    print("No se encontr√≥ una ra√≠z en el intervalo dado.")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gr√°fico de la funci√≥n y aproximaci√≥n de la ra√≠z usando bisecci√≥n en [3.2, 4]")
plt.legend()
plt.grid(True)
plt.show()
```

* Use el m√©todo de bisecci√≥n para encontrar una aproximaci√≥n dentro de 10^-5 para el primer valor positivo
de ùë• con ùë• = tan ùë•. 

```python
import numpy as np

# Definir la funci√≥n g(x) = x^2 - 1 - exp(1 - x^2)
def funcion_para_biseccion(valor_x):
    return valor_x**2 - 1 - np.exp(1 - valor_x**2)

# Implementaci√≥n modificada del m√©todo de Bisecci√≥n con precisi√≥n de 10^-3
def metodo_biseccion_ajustado(funcion, limite_inicial, limite_final, tolerancia=1e-3, max_iter=100):
    if funcion(limite_inicial) * funcion(limite_final) >= 0:
        print("No se puede aplicar el m√©todo de bisecci√≥n en este intervalo.")
        return None
    
    contador_iter = 0
    while (limite_final - limite_inicial) / 2 > tolerancia and contador_iter < max_iter:
        punto_central = (limite_inicial + limite_final) / 2
        valor_central = funcion(punto_central)
        
        if valor_central == 0:
            return punto_central  # Soluci√≥n exacta encontrada
        elif funcion(limite_inicial) * valor_central < 0:
            limite_final = punto_central
        else:
            limite_inicial = punto_central
        contador_iter += 1
    
    return (limite_inicial + limite_final) / 2  # Retorna la soluci√≥n aproximada

# Intervalo de b√∫squeda
inicio_intervalo, fin_intervalo = -2, 0

# Ejecutar el m√©todo de bisecci√≥n en el intervalo con la precisi√≥n ajustada
raiz_aproximada = metodo_biseccion_ajustado(funcion_para_biseccion, inicio_intervalo, fin_intervalo)
print("Aproximaci√≥n de la ra√≠z:", raiz_aproximada)
```

* Un abrevadero de longitud ùêø tiene una secci√≥n transversal en forma de semic√≠rculo con radio ùëü. (Consulte la
figura adjunta.) Cuando se llena con agua hasta una distancia ‚Ñé a partir de la parte superior, el volumen ùëâ de
agua es 

$$
V = L \left( 0.5 \pi r^2 - r^2 \arcsin\left( \frac{h}{r} \right) - h \left( r^2 - h^2 \right) \right)
$$

Suponga que $$ L = 10 \, \text{cm} $$, $$ r = 1 \, \text{cm} $$ y $$ V = 12.4 \, \text{cm}^3 $$. Encuentre la profundidad del agua en el abrevadero dentro de $$ 0.01 \, \text{cm} $$.




```python
import numpy as np
from scipy.optimize import root

# Par√°metros dados
longitud = 10  # Longitud en cm
radio = 1      # Radio en cm
volumen_deseado = 12.4  # Volumen en cm^3

# Definimos la funci√≥n que calcula el volumen en funci√≥n de h
def funcion_volumen(profundidad):
    if profundidad < 0 or profundidad > radio:
        return np.inf  # Retornar un valor muy alto si profundidad es inv√°lida

    # Calcula el volumen basado en la f√≥rmula dada
    componente1 = 0.5 * np.pi * radio**2
    componente2 = radio**2 * np.arcsin(profundidad / radio)
    componente3 = profundidad * np.sqrt(radio**2 - profundidad**2)
    volumen_calculado = longitud * (componente1 - componente2 - componente3)
    
    # La funci√≥n devuelve la diferencia entre el volumen deseado y el calculado
    return volumen_calculado - volumen_deseado

# Estimaci√≥n inicial para la profundidad (la mitad del radio est√° bien)
estimacion_inicial = radio / 2

# Usamos root para encontrar el valor de h
solucion = root(funcion_volumen, estimacion_inicial)

# Mostramos el resultado
if solucion.success:
    profundidad_agua = solucion.x[0]
    print(f"La profundidad del agua en el abrevadero es aproximadamente {profundidad_agua:.4f} cm")
else:
    print("No se pudo encontrar una soluci√≥n.")

```

* Un objeto que cae verticalmente a trav√©s del aire est√° sujeto a una resistencia viscosa, as√≠ como a la fuerza de gravedad. Suponga que un objeto con masa ùëö cae desde una altura 
 y que la altura del objeto despu√©s de ùë° segundos es:

 $$
s(t) = s_0 - \frac{L}{k} t + \frac{L}{k} \left(1 - e^{-kt}\right)
$$

Suponga que 

$$
g = 9.81 \, \text{m/s}^2 
$$

y $$ k $$ representa el coeficiente de la resistencia del aire en $$ \text{N s/m} $$. Suponga que $$ s_0 = 300 \, \text{m} $$, $$ m = 0.25 \, \text{kg} $$ y $$ k = 0.1 \, \text{N s/m} $$. Encuentre, dentro de $$ 0.01 \, \text{segundos} $$, el tiempo que tarda un cuarto de kg en golpear el piso.



```python
import numpy as np
from scipy.optimize import root

# Par√°metros dados
g = 9.81  # aceleraci√≥n de la gravedad en m/s^2
m = 0.25  # masa en kg
k = 0.1   # coeficiente de resistencia en Ns/m
s0 = 300  # altura inicial en metros

# Definimos la funci√≥n que calcula la altura en funci√≥n del tiempo
def altura(t):
    return s0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - np.exp(-k * t / m))

# Estimaci√≥n inicial para el tiempo (p.ej., 0.01 segundos)
estimacion_inicial = 0.01

# Usamos el m√©todo root para encontrar el tiempo que hace que la altura sea 0
solucion = root(altura, estimacion_inicial)

# Verificamos si se encontr√≥ una soluci√≥n
if solucion.success:
    tiempo_caida = solucion.x[0]
    print("El tiempo en que el objeto golpea el piso es aproximadamente:", tiempo_caida, "segundos")
else:
    print("No se pudo encontrar una soluci√≥n.")
```

### Ejercicio N¬∞1

* Use el teorema 2.1 para encontrar una cota para el n√∫mero de iteraciones necesarias para lograr una aproximaci√≥n con presici√≥n de 
 para la soluci√≥n de 
 que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximaci√≥n para la ra√≠z con este grado de precisi√≥n.

```python
import numpy as np
import matplotlib.pyplot as plt

class BisectionMethod:
    def __init__(self, func, a, b, epsilon):
        self.func = func
        self.a = a
        self.b = b
        self.epsilon = epsilon
        self.midpoints = []

    def calculate_iterations(self):
        N = np.ceil((np.log(self.b - self.a) - np.log(self.epsilon)) / np.log(2))
        return int(N)

    def find_root(self):
        if self.func(self.a) * self.func(self.b) >= 0:
            print("El m√©todo de bisecci√≥n no es aplicable en este intervalo.")
            return None
        
        iteration = 0
        while (self.b - self.a) / 2 > self.epsilon:
            midpoint = (self.a + self.b) / 2
            self.midpoints.append(midpoint)
            if self.func(midpoint) == 0:
                return midpoint  # Soluci√≥n exacta encontrada
            elif self.func(self.a) * self.func(midpoint) < 0:
                self.b = midpoint
            else:
                self.a = midpoint
            iteration += 1
            
        return (self.a + self.b) / 2  # Retorna la soluci√≥n aproximada

# Par√°metros del teorema
a, b = 1, 2
epsilon = 1e-4

# Definimos la funci√≥n para la ecuaci√≥n x^3 - x - 1 = 0
def f(x):
    return x**3 - x - 1

# Instanciamos la clase y calculamos el n√∫mero de iteraciones necesarias
bisection = BisectionMethod(f, a, b, epsilon)
print("N√∫mero m√≠nimo de iteraciones necesarias:", bisection.calculate_iterations())

# Ejecutar el m√©todo de bisecci√≥n
root_precise = bisection.find_root()
print("Aproximaci√≥n de la ra√≠z:", root_precise)

# Gr√°fica de la funci√≥n y los puntos de bisecci√≥n
x = np.linspace(0, 2, 100)  # Rango de x para la gr√°fica
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x¬≥ - x - 1', color='blue')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # L√≠nea horizontal en y=0
plt.axvline(root_precise, color='red', linestyle='--', label='Ra√≠z aproximada')
plt.scatter(bisection.midpoints, f(np.array(bisection.midpoints)), color='orange', label='Puntos medios', zorder=5)

# Etiquetas y t√≠tulo
plt.title('M√©todo de Bisecci√≥n')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-3, 3)  # Ajustar el rango del eje y para mejor visualizaci√≥n
plt.show()


```
