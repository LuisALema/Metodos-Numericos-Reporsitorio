#### 3. Diríjase al pseudocódigo del spline cúbico con frontera natural provisto en clase, en base a ese pseudocódigo complete la siguiente función:

![](imagen1.png)

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Función para calcular el error cuadrático medio
def calcular_mse(y_real, y_pred):
    return np.mean((y_real - y_pred)**2)

# Función para graficar los datos originales y el modelo ajustado
def graficar_ajuste(x, y, y_ajuste, titulo):
    plt.scatter(x, y, color='blue', label='Datos originales')
    plt.plot(x, y_ajuste, color='red', label='Ajuste')
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Ajuste polinomial de grado 1
coef1 = np.polyfit(xi, yi, 1)
ajuste1 = np.polyval(coef1, xi)
error1 = calcular_mse(yi, ajuste1)
print(f"Polinomio grado 1: coeficientes={coef1}, Error cuadrático medio={error1:.4f}")
graficar_ajuste(xi, yi, ajuste1, "Ajuste de polinomio grado 1")

# Ajuste polinomial de grado 2
coef2 = np.polyfit(xi, yi, 2)
ajuste2 = np.polyval(coef2, xi)
error2 = calcular_mse(yi, ajuste2)
print(f"Polinomio grado 2: coeficientes={coef2}, Error cuadrático medio={error2:.4f}")
graficar_ajuste(xi, yi, ajuste2, "Ajuste de polinomio grado 2")

# Ajuste polinomial de grado 3
coef3 = np.polyfit(xi, yi, 3)
ajuste3 = np.polyval(coef3, xi)
error3 = calcular_mse(yi, ajuste3)
print(f"Polinomio grado 3: coeficientes={coef3}, Error cuadrático medio={error3:.4f}")
graficar_ajuste(xi, yi, ajuste3, "Ajuste de polinomio grado 3")

# Ajuste de la forma b * exp(a * x)
log_yi = np.log(yi)
coef_exp = np.polyfit(xi, log_yi, 1)
a_exp, log_b_exp = coef_exp
b_exp = np.exp(log_b_exp)
ajuste_exp = b_exp * np.exp(a_exp * xi)
error_exp = calcular_mse(yi, ajuste_exp)
print(f"Ajuste exponencial (b*exp(a*x)): a={a_exp:.4f}, b={b_exp:.4f}, Error cuadrático medio={error_exp:.4f}")
graficar_ajuste(xi, yi, ajuste_exp, "Ajuste exponencial b*exp(a*x)")

# Ajuste de la forma b * x^a
log_xi = np.log(xi)
log_yi = np.log(yi)
coef_pot = np.polyfit(log_xi, log_yi, 1)
a_pot, log_b_pot = coef_pot
b_pot = np.exp(log_b_pot)
ajuste_pot = b_pot * xi**a_pot
error_pot = calcular_mse(yi, ajuste_pot)
print(f"Ajuste potencial (b*x^a): a={a_pot:.4f}, b={b_pot:.4f}, Error cuadrático medio={error_pot:.4f}")
graficar_ajuste(xi, yi, ajuste_pot, "Ajuste potencial b*x^a")

```
