```python
import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Función para calcular el error cuadrático medio
def calcular_mse(y_real, y_pred):
    return np.mean((y_real - y_pred)**2)

# Función para graficar los datos originales y el modelo ajustado
def graficar_ajuste(x, y, y_ajuste, titulo):
    plt.scatter(x, y, color='blue', label='Datos originales')
    plt.plot(x, y_ajuste, color='red', label='Ajuste')
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Ajuste polinomial de grado 1
coef1 = np.polyfit(xi, yi, 1)
ajuste1 = np.polyval(coef1, xi)
error1 = calcular_mse(yi, ajuste1)
print(f"Polinomio grado 1: coeficientes={coef1}, Error cuadrático medio={error1:.4f}")
graficar_ajuste(xi, yi, ajuste1, "Ajuste de polinomio grado 1")

# Ajuste polinomial de grado 2
coef2 = np.polyfit(xi, yi, 2)
ajuste2 = np.polyval(coef2, xi)
error2 = calcular_mse(yi, ajuste2)
print(f"Polinomio grado 2: coeficientes={coef2}, Error cuadrático medio={error2:.4f}")
graficar_ajuste(xi, yi, ajuste2, "Ajuste de polinomio grado 2")

# Ajuste polinomial de grado 3
coef3 = np.polyfit(xi, yi, 3)
ajuste3 = np.polyval(coef3, xi)
error3 = calcular_mse(yi, ajuste3)
print(f"Polinomio grado 3: coeficientes={coef3}, Error cuadrático medio={error3:.4f}")
graficar_ajuste(xi, yi, ajuste3, "Ajuste de polinomio grado 3")

# Ajuste de la forma b * exp(a * x)
log_yi = np.log(yi)
coef_exp = np.polyfit(xi, log_yi, 1)
a_exp, log_b_exp = coef_exp
b_exp = np.exp(log_b_exp)
ajuste_exp = b_exp * np.exp(a_exp * xi)
error_exp = calcular_mse(yi, ajuste_exp)
print(f"Ajuste exponencial (b*exp(a*x)): a={a_exp:.4f}, b={b_exp:.4f}, Error cuadrático medio={error_exp:.4f}")
graficar_ajuste(xi, yi, ajuste_exp, "Ajuste exponencial b*exp(a*x)")

# Ajuste de la forma b * x^a
log_xi = np.log(xi)
log_yi = np.log(yi)
coef_pot = np.polyfit(log_xi, log_yi, 1)
a_pot, log_b_pot = coef_pot
b_pot = np.exp(log_b_pot)
ajuste_pot = b_pot * xi**a_pot
error_pot = calcular_mse(yi, ajuste_pot)
print(f"Ajuste potencial (b*x^a): a={a_pot:.4f}, b={b_pot:.4f}, Error cuadrático medio={error_pot:.4f}")
graficar_ajuste(xi, yi, ajuste_pot, "Ajuste potencial b*x^a")

```

#### 3.La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos

![](imagen3.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales de puntuación ACT y promedios de puntos
act_scores = [28, 29, 25, 28, 28, 27, 27, 29, 28, 21, 33, 28, 28, 28, 29, 26, 23, 30, 27, 24]
gpa_scores = [3.84, 3.75, 3.21, 3.65, 3.23, 3.87, 3.63, 3.75, 3.75, 1.66, 3.20, 3.12, 3.41, 2.96, 3.38, 2.92, 3.53, 3.10, 2.03, 2.81]

# Ajuste lineal para encontrar la recta (mínimos cuadrados)
pendiente, intercepto = np.polyfit(act_scores, gpa_scores, 1)  # Calcula pendiente y el intercepto
print(f"La ecuación de la recta es: y = {pendiente:.2f}x + {intercepto:.2f}")

# Cálculo de los valores ajustados para la recta
recta_ajustada = [pendiente * x + intercepto for x in act_scores]

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.scatter(act_scores, gpa_scores, color='blue', label='Datos originales')  # Puntos originales
plt.plot(act_scores, recta_ajustada, color='red', label=f'Ajuste: y = {pendiente:.2f}x + {intercepto:.2f}')  # Recta ajustada
plt.title('Relación entre Puntuación ACT y Promedio de Puntos')
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de Puntos')
plt.legend()
plt.grid()
plt.show()


```

#### 4.El siguiente conjunto de datos, presentado al Subcomité Antimonopolio del Senado, muestra las características comparativas de supervivencia durante un choque de automóviles de diferentes clases. Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria). 


![](imagen4.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de peso promedio y porcentaje de presentación
pesos = [4800, 3700, 3400, 2800, 1900]  # En libras
porcentajes = [3.1, 4.0, 5.2, 6.4, 9.6]  # En porcentaje

# Calcular la recta de ajuste (mínimos cuadrados)
m, b = np.polyfit(pesos, porcentajes, 1)  # Calcula pendiente y el intercepto
print(f"Ecuación de la recta: y = {m:.2f}x + {b:.2f}")

# Generar los valores ajustados con la ecuación de la recta
ajuste = [m * peso + b for peso in pesos]

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.scatter(pesos, porcentajes, color='blue', label='Datos originales')  # Puntos originales
plt.plot(pesos, ajuste, color='red', label=f'Ajuste: y = {m:.2f}x + {b:.2f}')  # Línea ajustada
plt.title('Relación entre Peso Promedio y Porcentaje de Presentación')
plt.xlabel('Peso promedio (libras)')
plt.ylabel('Porcentaje de presentación')
plt.legend()
plt.grid()
plt.show()
```
