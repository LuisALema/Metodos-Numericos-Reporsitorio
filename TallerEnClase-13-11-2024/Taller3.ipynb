### 1. GRAFICA DE LA FUNCION COSENO. -


```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el rango de valores para la variable independiente
valores_x = np.linspace(-8, 8, 400)

# Aproximaciones de Taylor para cos(x) con distintos términos
aprox_cos_2 = 1 - (valores_x**2) / 2                 # Aproximación con 2 términos
aprox_cos_4 = aprox_cos_2 + (valores_x**4) / 24       # Aproximación con 4 términos
aprox_cos_6 = aprox_cos_4 - (valores_x**6) / 720      # Aproximación con 6 términos

# Crear una figura y gráficos
plt.figure(figsize=(10, 6))
plt.plot(valores_x, np.cos(valores_x), label='cos(x)', color='blue')  # Función coseno original
plt.plot(valores_x, aprox_cos_2, label='2 términos', color='orange', linestyle='--')
plt.plot(valores_x, aprox_cos_4, label='4 términos', color='green', linestyle='--')
plt.plot(valores_x, aprox_cos_6, label='6 términos', color='red', linestyle='--')

# Etiquetas y personalización del gráfico
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Serie de Taylor para cos(x) con varias aproximaciones")
plt.legend()
plt.grid(True)
plt.xlim(-8, 8)
plt.ylim(-5, 5)
plt.show()

```

### 2. GRAFICA DE LA FUNCION $1/(1-X)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x, evitando valores donde la función diverge
valores_x = np.linspace(-0.9, 0.9, 400)

# Aproximaciones de Taylor para 1 / (1 - x) usando diferentes cantidades de términos
taylor_aprox_1 = 1
taylor_aprox_2 = taylor_aprox_1 + valores_x
taylor_aprox_3 = taylor_aprox_2 + valores_x**2
taylor_aprox_4 = taylor_aprox_3 + valores_x**3
taylor_aprox_5 = taylor_aprox_4 + valores_x**4
taylor_aprox_6 = taylor_aprox_5 + valores_x**5

# Generar la gráfica para la función original y sus aproximaciones
plt.figure(figsize=(10, 6))
plt.plot(valores_x, 1 / (1 - valores_x), label=r'$\frac{1}{1 - x}$', color='blue')  # Función original
plt.plot(valores_x, taylor_aprox_1 * np.ones_like(valores_x), label='1 término: 1', color='orange', linestyle='--')
plt.plot(valores_x, taylor_aprox_2, label='2 términos: 1 + x', color='green', linestyle='--')
plt.plot(valores_x, taylor_aprox_3, label='3 términos: 1 + x + x²', color='red', linestyle='--')
plt.plot(valores_x, taylor_aprox_4, label='4 términos: 1 + x + x² + x³', color='purple', linestyle='--')
plt.plot(valores_x, taylor_aprox_5, label='5 términos: 1 + x + x² + x³ + x⁴', color='brown', linestyle='--')
plt.plot(valores_x, taylor_aprox_6, label='6 términos: 1 + x + x² + x³ + x⁴ + x⁵', color='pink', linestyle='--')

# Configuración del gráfico
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximaciones de Taylor de $\\frac{1}{1 - x}$ con varios términos")
plt.legend()
plt.grid(True)
plt.ylim(-1, 5)
plt.show()


```

### 3. GRAFICA DE LA FUNCION $ln(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función principal
def funcion_logaritmica(x):
    return np.log(x)

# Definir el polinomio de Taylor de tercer grado
def polinomio_taylor_3er_grado(x):
    delta_x = x - 1
    return delta_x - (delta_x**2) / 2 + (delta_x**3) / 3

# Rango de valores de x
rango_x = np.linspace(0.1, 2, 400)
valores_funcion = funcion_logaritmica(rango_x)
valores_taylor = polinomio_taylor_3er_grado(rango_x)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(rango_x, valores_funcion, label=r'$f(x) = \ln(x)$', color='blue')
plt.plot(rango_x, valores_taylor, label='Polinomio de Taylor de tercer grado', color='orange', linestyle='--')

# Configuración del gráfico
plt.title('Comparación entre la función $\ln(x)$ y su polinomio de Taylor de tercer grado')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.ylim(-2, 2)
plt.xlim(0, 2)
plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
